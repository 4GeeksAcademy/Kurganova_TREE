```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
import numpy as np  # Agregar esta línea para importar NumPy

# Definir los hiperparámetros que deseas optimizar
param_grid = {
    'n_estimators': [int(x) for x in np.linspace(start = 50, stop = 200, num = 10)], # Número de árboles en el bosque
    'max_features': ['auto', 'sqrt'], # Número máximo de características consideradas para dividir un nodo
    'max_depth': [int(x) for x in np.linspace(10, 110, num = 11)], # Profundidad máxima del árbol
    'min_samples_split': [2, 5, 10], # Número mínimo de muestras requeridas para dividir un nodo
    'min_samples_leaf': [1, 2, 4], # Número mínimo de muestras requeridas en cada nodo hoja
    'bootstrap': [True, False] # Método de muestreo de muestras
}

# Crear el modelo base
rf = RandomForestClassifier(random_state=42)

# Realizar la búsqueda aleatoria de hiperparámetros
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = param_grid, 
                               n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)

# Ajustar el modelo a los datos
rf_random.fit(X_train, y_train)

# Ver los mejores hiperparámetros encontrados
print(rf_random.best_params_)

# Utilizar el modelo optimizado para predecir
y_pred = rf_random.best_estimator_.predict(X_test)
```
